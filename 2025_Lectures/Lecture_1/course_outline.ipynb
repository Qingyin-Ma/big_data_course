### Jupyter Notebook Shortcuts

#### Command Mode Cell (blue border)

- `A`: create empty cell above


- `B`: create empty cell below


- `X`: cut the selected cell


- `V`: paste the copied cell below


- `DD`: delete the selected cell


- `M`: switch to Markdown mode


- `Y`: switch to Code mode


- `Enter`: switch from command mode to edit mode



#### Edit Mode Cell (green border)

- `Esc`: switch from edit mode to command mode

- `Ctrl + /`: comment the code (or uncomment the commented code)


#### Command or Edit Mode Cell

- `Shift + Enter`: run the cell


- `Ctrl + S`: save the curent notebook

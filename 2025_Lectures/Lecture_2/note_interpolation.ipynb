## Fast Interpolation

#### Reference: 
https://www.econforge.org/interpolation.py/

```python
import numpy as np
import matplotlib.pyplot as plt
from interpolation import interp
```

```python
f = lambda x: -2*(x**2) + 3*x + 1
x_in = np.linspace(-10, 10, 200)
f_interp = lambda x: interp(x_in, f(x_in), x)

fig, ax = plt.subplots()
x_vals = np.linspace(-15, 15, 200)
ax.plot(x_vals, f(x_vals), label='real function value')
ax.plot(x_vals, f_interp(x_vals), ls='--', lw=2.5, label='default interpolation')
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$', rotation=0, labelpad=10)
ax.legend()
plt.title('Demonstrating interpolation: $f(x) = -2x^2 + 3x + 1$')
plt.show()
```
